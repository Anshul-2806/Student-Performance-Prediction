```python
# PART 1 : DATA LOADING + EXPLORATORY DATA ANALYSIS (EDA)
# ==========================================================

# ==========================================================
# STEP 1 : Import Required Libraries
# ==========================================================

# Pandas -> Dataset ko read aur manipulate karne ke liye
import pandas as pd

# NumPy -> Mathematical operations ke liye
import numpy as np

# Matplotlib -> Basic graphs banane ke liye
import matplotlib.pyplot as plt

# Seaborn -> Beautiful visualization ke liye
import seaborn as sns

# Warning hide karne ke liye
import warnings
warnings.filterwarnings("ignore")

# Graph style
sns.set_style("whitegrid")

print("All Libraries Imported Successfully")


# ==========================================================
# STEP 2 : Load Dataset
# ==========================================================

# Dataset load kar rahe hain

df = pd.read_csv("Student_Marks.csv")

print("\nDataset Loaded Successfully")


# ==========================================================
# STEP 3 : First Five Rows
# ==========================================================

print("\nFirst 5 Rows")

print(df.head())


# ==========================================================
# STEP 4 : Last Five Rows
# ==========================================================

print("\nLast 5 Rows")

print(df.tail())


# ==========================================================
# STEP 5 : Dataset Shape
# ==========================================================

print("\nDataset Shape")

print(df.shape)

print("\nRows :",df.shape[0])

print("Columns :",df.shape[1])


# ==========================================================
# STEP 6 : Column Names
# ==========================================================

print("\nColumn Names")

print(df.columns)


# ==========================================================
# STEP 7 : Dataset Information
# ==========================================================

print("\nDataset Information")

df.info()


# ==========================================================
# STEP 8 : Statistical Summary
# ==========================================================

print("\nStatistical Summary")

print(df.describe())


# ==========================================================
# STEP 9 : Missing Values
# ==========================================================

print("\nMissing Values")

print(df.isnull().sum())


# ==========================================================
# STEP 10 : Duplicate Records
# ==========================================================

print("\nDuplicate Rows")

print(df.duplicated().sum())


# ==========================================================
# STEP 11 : Remove Duplicate Rows
# ==========================================================

df = df.drop_duplicates()

print("\nShape After Removing Duplicates")

print(df.shape)


# ==========================================================
# STEP 12 : Check Data Types
# ==========================================================

print("\nData Types")

print(df.dtypes)


# ==========================================================
# STEP 13 : Correlation Matrix
# ==========================================================

print("\nCorrelation")

print(df.corr())


# ==========================================================
# GRAPH 1 : Correlation Heatmap
# ==========================================================

plt.figure(figsize=(6,5))

sns.heatmap(

    df.corr(),

    annot=True,

    cmap="coolwarm"

)

plt.title("Correlation Heatmap")

plt.show()


# ==========================================================
# GRAPH 2 : Distribution of Marks
# ==========================================================

plt.figure(figsize=(7,5))

sns.histplot(

    df["Marks"],

    bins=10,

    kde=True

)

plt.title("Distribution of Marks")

plt.show()


# ==========================================================
# GRAPH 3 : Distribution of Study Time
# ==========================================================

plt.figure(figsize=(7,5))

sns.histplot(

    df["time_study"],

    bins=10,

    kde=True,

    color="green"

)

plt.title("Study Time Distribution")

plt.show()


# ==========================================================
# GRAPH 4 : Courses Distribution
# ==========================================================

plt.figure(figsize=(7,5))

sns.countplot(

    x=df["number_courses"]

)

plt.title("Number of Courses")

plt.show()


# ==========================================================
# GRAPH 5 : Marks vs Study Time
# ==========================================================

plt.figure(figsize=(8,6))

sns.scatterplot(

    x=df["time_study"],

    y=df["Marks"],

    s=100

)

plt.title("Study Time vs Marks")

plt.show()


# ==========================================================
# GRAPH 6 : Courses vs Marks
# ==========================================================

plt.figure(figsize=(8,6))

sns.scatterplot(

    x=df["number_courses"],

    y=df["Marks"],

    s=100,

    color="red"

)

plt.title("Courses vs Marks")

plt.show()


# ==========================================================
# GRAPH 7 : Boxplot
# ==========================================================

plt.figure(figsize=(6,5))

sns.boxplot(

    y=df["Marks"]

)

plt.title("Marks Boxplot")

plt.show()


# ==========================================================
# GRAPH 8 : Pairplot
# ==========================================================

sns.pairplot(df)

plt.show()


# ==========================================================
# GRAPH 9 : Study Time Boxplot
# ==========================================================

plt.figure(figsize=(6,5))

sns.boxplot(

    y=df["time_study"],

    color="orange"

)

plt.show()


# ==========================================================
# GRAPH 10 : Courses Pie Chart
# ==========================================================

course_count = df["number_courses"].value_counts()

plt.figure(figsize=(7,7))

plt.pie(

    course_count,

    labels=course_count.index,

    autopct="%1.1f%%"

)

plt.title("Courses Distribution")

plt.show()


# ==========================================================
# STEP 14 : Feature Selection
# ==========================================================

# Independent Variables

X = df[["number_courses","time_study"]]

# Dependent Variable

y = df["Marks"]


print("\nFeatures")

print(X.head())

print("\nTarget")

print(y.head())


# ==========================================================
# STEP 15 : Feature Scaling Required?
# ==========================================================

print("\nChecking Feature Range")

print(X.describe())


# ==========================================================
# STEP 16 : Final Clean Dataset
# ==========================================================

print("\nFinal Dataset Ready")

print(df.head())

print("\nDataset Shape :",df.shape)


# ==========================================================
# PART 1 COMPLETED
# ==========================================================

print("\nEDA Completed Successfully")
```

```python
# ==========================================================
# STUDENT PERFORMANCE PREDICTION USING MACHINE LEARNING
# PART 2 : MODEL TRAINING & EVALUATION
# ==========================================================

# ==========================================================
# STEP 1 : Import Required Libraries
# ==========================================================

# Train Test Split
from sklearn.model_selection import train_test_split

# Regression Models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Evaluation Metrics
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# Save Model
import joblib

# Visualization
import matplotlib.pyplot as plt

import numpy as np

print("Libraries Imported Successfully")


# ==========================================================
# STEP 2 : Train Test Split
# ==========================================================

# 80% data training ke liye
# 20% testing ke liye

X_train, X_test, y_train, y_test = train_test_split(

    X,
    y,

    test_size=0.20,

    random_state=42

)

print("Training Samples :",len(X_train))

print("Testing Samples :",len(X_test))


# ==========================================================
# STEP 3 : Linear Regression
# ==========================================================

print("\nTraining Linear Regression...")

linear_model = LinearRegression()

linear_model.fit(

    X_train,
    y_train

)

linear_prediction = linear_model.predict(X_test)

print("Linear Regression Completed")


# ==========================================================
# STEP 4 : Decision Tree Regression
# ==========================================================

print("\nTraining Decision Tree...")

tree_model = DecisionTreeRegressor(

    random_state=42

)

tree_model.fit(

    X_train,
    y_train

)

tree_prediction = tree_model.predict(X_test)

print("Decision Tree Completed")


# ==========================================================
# STEP 5 : Random Forest Regression
# ==========================================================

print("\nTraining Random Forest...")

forest_model = RandomForestRegressor(

    n_estimators=100,

    random_state=42

)

forest_model.fit(

    X_train,
    y_train

)

forest_prediction = forest_model.predict(X_test)

print("Random Forest Completed")


# ==========================================================
# STEP 6 : Function for Evaluation
# ==========================================================

def evaluate_model(name, actual, prediction):

    mae = mean_absolute_error(actual, prediction)

    mse = mean_squared_error(actual, prediction)

    rmse = np.sqrt(mse)

    r2 = r2_score(actual, prediction)

    print("\n===========================")

    print(name)

    print("===========================")

    print("MAE :", round(mae,2))

    print("MSE :", round(mse,2))

    print("RMSE :", round(rmse,2))

    print("R2 Score :", round(r2,4))

    return r2


# ==========================================================
# STEP 7 : Evaluate All Models
# ==========================================================

linear_score = evaluate_model(

    "Linear Regression",

    y_test,

    linear_prediction

)

tree_score = evaluate_model(

    "Decision Tree",

    y_test,

    tree_prediction

)

forest_score = evaluate_model(

    "Random Forest",

    y_test,

    forest_prediction

)


# ==========================================================
# STEP 8 : Model Comparison
# ==========================================================

models = [

    "Linear",

    "Decision Tree",

    "Random Forest"

]

scores = [

    linear_score,

    tree_score,

    forest_score

]

plt.figure(figsize=(8,5))

plt.bar(

    models,

    scores

)

plt.title("Model Comparison (R² Score)")

plt.ylabel("R² Score")

plt.show()


# ==========================================================
# STEP 9 : Best Model Selection
# ==========================================================

best_score = max(scores)

best_model = None

if best_score == linear_score:

    best_model = linear_model

    print("\nBest Model : Linear Regression")

elif best_score == tree_score:

    best_model = tree_model

    print("\nBest Model : Decision Tree")

else:

    best_model = forest_model

    print("\nBest Model : Random Forest")


# ==========================================================
# STEP 10 : Save Best Model
# ==========================================================

joblib.dump(

    best_model,

    "student_model.pkl"

)

print("\nModel Saved Successfully")


# ==========================================================
# STEP 11 : Predict New Student
# ==========================================================

print("\n========== Prediction ==========")

number_courses = 6

study_time = 5.5

prediction = best_model.predict(

    [[number_courses, study_time]]

)

print("Courses :",number_courses)

print("Study Hours :",study_time)

print("Predicted Marks :",round(prediction[0],2))


# ==========================================================
# STEP 12 : Actual vs Predicted Graph
# ==========================================================

plt.figure(figsize=(8,5))

plt.plot(

    y_test.values,

    label="Actual",

    marker='o'

)

plt.plot(

    best_model.predict(X_test),

    label="Predicted",

    marker='x'

)

plt.title("Actual vs Predicted Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.legend()

plt.show()


# ==========================================================
# STEP 13 : Scatter Plot
# ==========================================================

plt.figure(figsize=(6,6))

plt.scatter(

    y_test,

    best_model.predict(X_test)

)

plt.xlabel("Actual Marks")

plt.ylabel("Predicted Marks")

plt.title("Prediction Accuracy")

plt.show()


# ==========================================================
# STEP 14 : Feature Importance
# ==========================================================

# Random Forest aur Decision Tree me feature importance hoti hai

if hasattr(best_model, "feature_importances_"):

    importance = best_model.feature_importances_

    features = X.columns

    plt.figure(figsize=(6,5))

    plt.bar(

        features,

        importance

    )

    plt.title("Feature Importance")

    plt.show()

else:

    print("\nLinear Regression me Feature Importance available nahi hoti.")


# ==========================================================
# STEP 15 : Load Saved Model
# ==========================================================

loaded_model = joblib.load(

    "student_model.pkl"

)

print("\nSaved Model Loaded Successfully")


# ==========================================================
# STEP 16 : Predict Using Loaded Model
# ==========================================================

result = loaded_model.predict(

    [[4,6]]

)

print("\nPrediction Using Saved Model")

print("Predicted Marks :",round(result[0],2))


# ==========================================================
# PROJECT COMPLETED
# ==========================================================

print("\n========================================")

print("Student Performance Prediction Completed")

print("========================================")
```

```python
# ==========================================================
# STUDENT PERFORMANCE PREDICTION WEB APP
# ==========================================================

import streamlit as st
import joblib

# Load trained model
model = joblib.load("student_model.pkl")

# Page Configuration
st.set_page_config(
    page_title="Student Performance Prediction",
    page_icon="📚",
    layout="centered"
)

# Title
st.title("📚 Student Performance Prediction")

st.write(
    "Predict student marks using a trained Machine Learning Regression Model."
)

st.markdown("---")

# User Inputs
number_courses = st.slider(
    "Number of Courses",
    1,
    10,
    5
)

time_study = st.slider(
    "Study Time (Hours)",
    1.0,
    10.0,
    5.0
)

# Prediction
if st.button("Predict Marks"):

    prediction = model.predict([[number_courses, time_study]])

    st.success(
        f"Predicted Marks : {prediction[0]:.2f}"
    )

st.markdown("---")

st.caption("Developed using Python, Scikit-Learn & Streamlit")
```
